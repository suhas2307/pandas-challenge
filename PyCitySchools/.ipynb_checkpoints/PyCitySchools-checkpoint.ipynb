```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate total schools by finding unique school names and taking the length of the array object
total_schools = len(school_data_complete["school_name"].unique())

# Calculate total students by counting the Student IDs which are unique to each student
total_students = school_data_complete['Student ID'].count()

# School budget is repeated for each student of the same school. 
# To calculate total budget, create a dataframe with unique values for School and Budget
budget_df = school_data_complete[["school_name","budget"]].drop_duplicates()
total_budget = budget_df['budget'].sum()

# Calculating average math score
avg_math_score = round(school_data_complete["math_score"].mean(),6)

# Calculating average reading score
avg_reading_score = round(school_data_complete["reading_score"].mean(),5)

# Calculating overall passing rate
overall_passing_rate = round((avg_math_score + avg_reading_score)/2,6)

# To calculate % of students with a passing math score, get the rows of the dataframe having math score >=70
# Then calculate the count of the no.of rows
tot_stu_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()

# To calculate % of students with a passing math score, get the rows of the dataframe having math score >=70
# Then calculate the count of the no.of rows
tot_stu_pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["math_score"].count()

# Creating Dataframe
district_summary_df = pd.DataFrame([{"Total Schools":total_schools,"Total Students":f"{total_students:,}",
              "Total Budget":f"${total_budget:,.2f}","Average Math Score":avg_math_score,
              "Average Reading Score":avg_reading_score,
               "% Passing Math":round((tot_stu_pass_math/total_students)*100,6),
              "% Passing Reading":round((tot_stu_pass_read/total_students)*100,6),
              "% Overall Passing Rate":overall_passing_rate}])
district_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Create a Dataframe containing the school type and budget
# Set Index as school_name
school_type_budget_df = school_data_complete[["school_name","type","budget"]].drop_duplicates()
school_type_budget_df.set_index(keys="school_name",inplace=True)

# Get total students in each school
school_groupby = school_data_complete.groupby("school_name")
stu_count_by_school = school_groupby["Student ID"].count()

# Get per student budget by dividing total budget by no.of students in the school
stu_budget = school_type_budget_df["budget"]/stu_count_by_school

# Get avg math score for the school
school_avg_math = school_groupby["math_score"].mean()

# Get avg reading score for the school
school_avg_read = school_groupby["reading_score"].mean()

# Get percentage passing math in each school
stu_math_pass_per = school_data_complete.loc[school_data_complete["math_score"]>=70]["school_name"].value_counts()/stu_count_by_school*100

# Get percentage passing reading in each school
stu_read_pass_per = school_data_complete.loc[school_data_complete["reading_score"]>=70]["school_name"].value_counts()/stu_count_by_school*100

# Get Overall Passing rate
stu_overall_pass_per = (stu_math_pass_per + stu_read_pass_per)/2

columns = ["School Type","Total Students","Total School Budget","Per Student Budget","Average Math Score",
          "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]

#Create an empty DataFrame
school_summary_df = pd.DataFrame(columns=columns)
# Add column data to the Dataframe from previously created series
school_summary_df["School Type"] = school_type_budget_df["type"]
school_summary_df["Total Students"] = stu_count_by_school
school_summary_df["Total School Budget"] = school_type_budget_df["budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = stu_budget.map("${:.2f}".format)
school_summary_df["Average Math Score"] = school_avg_math
school_summary_df["Average Reading Score"] = school_avg_read
school_summary_df["% Passing Math"] = stu_math_pass_per
school_summary_df["% Passing Reading"] = stu_read_pass_per
school_summary_df["% Overall Passing Rate"] = stu_overall_pass_per
#Sort the dataframe in descending order
school_summary_df.sort_values("% Overall Passing Rate",ascending=False).iloc[0:5]
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
school_summary_df.sort_values("% Overall Passing Rate",ascending=True).iloc[0:5]
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a series for the math average for each grade
grade_9 = school_data_complete.loc[school_data_complete["grade"]=="9th"]
grade_9_math_avg = grade_9.groupby("school_name")["math_score"].mean()

grade_10 = school_data_complete.loc[school_data_complete["grade"]=="10th"]
grade_10_math_avg = grade_10.groupby("school_name")["math_score"].mean()

grade_11 = school_data_complete.loc[school_data_complete["grade"]=="11th"]
grade_11_math_avg = grade_11.groupby("school_name")["math_score"].mean()

grade_12 = school_data_complete.loc[school_data_complete["grade"]=="12th"]
grade_12_math_avg = grade_12.groupby("school_name")["math_score"].mean()

# Creating a Dataframe to hold the results
columns = ["9th","10th","11th","12th"]
math_scores_by_grade = pd.DataFrame({"9th":grade_9_math_avg,"10th":grade_10_math_avg,"11th":grade_11_math_avg,"12th":grade_12_math_avg})
math_scores_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# Calculate average reading score by grade
grade_9_read_avg = grade_9.groupby("school_name")["reading_score"].mean()
grade_10_read_avg = grade_10.groupby("school_name")["reading_score"].mean()
grade_11_read_avg = grade_11.groupby("school_name")["reading_score"].mean()
grade_12_read_avg = grade_12.groupby("school_name")["reading_score"].mean()

# Creating a Dataframe to hold the results
read_scores_by_grade = pd.DataFrame({"9th":grade_9_read_avg,"10th":grade_10_read_avg,"11th":grade_11_read_avg,"12th":grade_12_read_avg})
read_scores_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Convert Per Student Budget series into Integer for binning
per_student_budget = school_summary_df["Per Student Budget"].replace('\$',"",regex=True).astype("float").astype("int")
school_summary_df["Budget Summary"] = pd.cut(per_student_budget,bins=spending_bins,labels=group_names,include_lowest=True)
school_summary_df.groupby("Budget Summary")[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_summary_df["School Size Summary"] = pd.cut(school_summary_df["Total Students"],bins=size_bins,labels=group_names)
school_summary_df.groupby("School Size Summary")[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]].mean()
```
